### 完成迭代协议

```python
L = [1, 2, 3]
print(dir(L))
# __iter__ 
```

```python
I = iter(L) # 其实是调用其自身的 __iter__方法
print(type(I))
print(dir(I))
# __iter__ __next__
```

```python
I.__next__(), next(I), next(I)
```

```python
# 迭代器迭代完了, 不能再次迭代
I.__next__()
```

## 列表推导:初次深入探索

```python
L = [1, 2, 3, 4, 5]
print(id(L))

for i in range(len(L)):
    L[i] += 10
print(id(L))
print(L) # L 还是原来的L
```

```python
L = [x + 10 for x in L]
print(id(L)) # 不是原来的L了
print(L)
```

## 多遍迭代器 vs 单遍迭代器

```python
r = range(5)
i1 = iter(r)
i2 = iter(r)
print(i1 is i2) # 两个迭代器
next(i1), next(i1), next(i2), next(i2)
```

```python
z = zip((1, 2, 3, 4), (10, 11, 12, 13))
i1 = iter(z)
i2 = iter(z)
print(i1 is i2) # 同一个迭代器
next(i1), next(i1), next(i2), next(i2)
```

```python
m = map(abs, [-1, -2, 3, 4])
i1 = iter(m)
i2 = iter(m)
print(i1 is i2) # 同一个迭代器
next(i1), next(i1), next(i2), next(i2)
```

```python
r = range(5)
i1 = iter(r)
i2 = iter(r)
print(i1 is i2) # 同一个迭代器
next(i1), next(i1), next(i2), next(i2)
```
