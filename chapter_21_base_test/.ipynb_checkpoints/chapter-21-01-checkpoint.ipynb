## timeit

```python
%%timeit
[x ** 2 for x in range(1000)]
```

```python
import timeit
```

```python
timeit.repeat(stmt="[x ** 2 for x in range(1000)] ", number=1000, repeat=5) 
```

```python
import time
s = time.time()
n = 1000
for i in range(n):
    [x ** 10 for x in range(100000)]
(time.time() - s) / n * 1000
```
