## 参数传递基础

```python
def f(a):
    a = 99
    
b = 88
f(b)
print(b)
```

```python
def changer(a, b):
    b[0] = 'spam'
    
X = 1
L = [1, 2]
changer(X, L)
X, L
```

```python
X = 1
L = [1, 2]
changer(X, L[:]) # 使用复制的内容,原始内容不变
X, L
```

### 模拟输出参数和多重结果

```python
def multiple(x, y):
    x = 2
    y = [3, 4]
    return x, y

X = 1
L = [1, 2]
X, L = multiple(X, L)
X, L
```

### 可变长参数的实例

#### 函数定义中:收集参数

```python
def f(*args):
    print(args)
```

```python
f()
```

```python
f(1)
```

```python
f(1, 2, 3, 4)
```

```python
def f(**args):
    print(type(args))
    print(args)
```

```python
f()
```

```python
f(a=1, b=2)
```

```python
def f3(a, *pargs, **kargs):
    print(a, pargs, kargs)
```

```python
f3(1, 2, 3, x=1, y=2)
```

```python
f3(1, 2, 3)
```

#### 函数调用中 解包参数

```python
def func2(a, b, c, d):
    print(a, b, c, d)
```

```python
args = (1, 2)
args += (3, 4)
args
```

```python
func2(*args)
```

```python
args = {"a": 1, "b":2, "c": 3, "d": 4}
func2(**args)
```

```python
args = {"a": 1, "b":2, "c": 3}
func2(**args)
```

```python
args = {"a": 1, "b":2, "c": 3, "d": 4, "e":5}
func2(**args)
```

```python
func2(*(1, 2), **{'d':4, 'c':3})
```

```python
func2(1, *(2, 3), **{'d':4})
```

```python
func2(1, c=3, *(2, ), **{'d': 4})
```

```python
func2(1, *(2, 3), d=4)
```

```python
func2(1, *(2,), c=3, **{'d':4})
```

```python
func2(1, *[2, 3], **{'d':4})
```

### 泛化地使用函数

```python
args = (2, 3, 4)
func2(args)
```

```python
def tracer(func, *pargs, **kargs):
    print('calling: ', func.__name__)
    return func(*pargs, **kargs)

def func(a, b, c, d):
    print(a + b + c + d)
    
tracer(func, 1, 2, c=3, d=4)
```

### 3.X 的keyword-only参数

```python
def kwonly(a, *b, c):
    print(a, b, c)
```

```python
kwonly(1, 2, c=3)
```

```python
kwonly(a=1, c=3)
```

```python
kwonly(1, c=3)
```

```python
kwonly(1, 2, 3) # 2, 3全给b了
```

```python
def kwonly(a, *, b=1, c, d=2):
    print(a, b, c, d)
```

```python
kwonly(3, c=4)
```

```python
kwonly(3, c=4, b=5)
```

```python
kwonly(3)
```

```python
kwonly(1, 2, 3)
```

```python
def kwonly(a, *, b=2, c=3, d=4):
    print(a, b, c, d)
    
kwonly(1, 2, 3, 4)
```

```python
def kwonly(a, b=2, c=3, d=4):
    print(a, b, c, d)
    
kwonly(1, 2, 3, 4)
```
