## 参数传递基础

```python
def f(a):
    a = 99
    
b = 88
f(b)
print(b)
```

```python
def changer(a, b):
    b[0] = 'spam'
    
X = 1
L = [1, 2]
changer(X, L)
X, L
```

```python
X = 1
L = [1, 2]
changer(X, L[:]) # 使用复制的内容,原始内容不变
X, L
```

### 模拟输出参数和多重结果

```python
def multiple(x, y):
    x = 2
    y = [3, 4]
    return x, y

X = 1
L = [1, 2]
X, L = multiple(X, L)
X, L
```

## min提神小案例

```python
def mini1(*args):
    return min(args)
```

```python
mini1(1, 2, 3, 4)
```

```python
mini1('a','T', '3')
```

```python
def min1(*args):
    res = args[0]
    for arg in args[1:]:
        if arg < res:
            res = arg
    return res
```

```python
def min2(first, *args):
    for arg in args:
        if first < arg:
            first = arg
    return first
```

```python
def min3(*args):
    tmp = list(args)
    tmp.sort()
    return tmp[0]
```

```python
print(min1(3, 4, 1, 2))
print(min2('bb', 'aa'))
print(min3([2, 3], [1, 1], [3, 3]))
```

### 附加分

```python
def minmax(test, *args):
    res = args[0]
    for arg in args[1:]:
        if test(arg, res):
            res = arg
    return res

def lessthan(x, y):
    return x < y

def greathan(x, y):
    return x > y
```

```python
minmax(lessthan, 4, 2, 1, 5, 6, 3)
```

```python
minmax(greathan, 4, 2, 1, 5, 6, 3)
```

## 通用set函数

```python
def intersect(*args):
    res = []
    for x in args[0]:
        if x in res:
            continue
        for other in args[1:]:
            if x not in other:
                break
        else:
            res.append(x)
    return res
```

```python
def union(*args):
    res = []
    for seq in args:
        for x in seq:
            if not x in res:
                res.append(x)
    return res
```

```python
s1 = 'spam'
s2 = 'scam'
s3 = 'slam'
intersect(s1, s2)
```

```python
intersect(s1, s2, s3)
```

```python
intersect([1, 2, 3], (1, 4))
```

```python
union(s1, s2)
```

```python
union(s1, s2, s3)
```

```python
def tester(func, items, trace=True):
    for i in range(len(items)):
        items = items[1:] + items[:1]
        if trace:
            print(items)
        print(sorted(func(*items)))
```

```python
tester(intersect, ('a', 'abcdefg', 'abdst', 'albmcnd'))
```

```python
tester(union, ('a', 'abcdefg', 'abdst', 'albmcnd'), False)
```

```python
tester(intersect, ('ba', 'abcdefg', 'abdst', 'albmcnd'), False)
```

```python
intersect([1, 2, 1, 3], (1, 1, 4))
```

```python
tester(intersect, ('ababa', 'abcdefg', 'aaaab'), False)
```
