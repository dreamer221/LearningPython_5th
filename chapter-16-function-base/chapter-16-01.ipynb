## 第一个示例: 定义和调用

```python
def times(x, y):
    return x * y
```

```python
times(2, 4)
```

```python
times(3.14, 4)
```

```python
times('Ni', 4)
```

## 第二个示例: 寻找序列的子集

```python
def intersect(seq1, seq2):
    res = []
    for x in seq1:
        if x in seq2:
            res.append(x)
    return res
```

```python
s1 = 'SPAM'
s2 = 'SCAM'
intersect(s1, s2)
```

```python
[x for x in s1 if x in s2]
```

```python
intersect([1, 2, 3], (1, 4))
```
