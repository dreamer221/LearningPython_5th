## json

```python
name = {'first': 'Bob', 'last': 'Smith'}
rec = dict(name=name, job=['dev', 'mgr'], age=40.5)
rec
```

```python
type(rec)
```

```python
import json

# dict --> str
s = json.dumps(rec)
type(s), s
```

```python
# str --> dict
d2 = json.loads(s)
type(d2), d2
```

```python
# dict --> file
f = open('testjson.txt', 'w')
json.dump(rec, f, indent=4)
f.close()
```

```python
# file -> str
s2 = open('testjson.txt', 'r').read()
print(type(s2))
print(s2)
```

```python
d = json.load(open('testjson.txt'))
type(d)
```

# 内置类型陷阱

```python
L = [1, 2, 3]
M = ['X', L, 'Y']
print(M)

L[1] = 666
L, M
```

```python
L = [1, 2, 3]
M = ['X', L[:], 'Y']
print(M)

L[1] = 666
L, M
```

## 重复会增加深度

```python
L = [4, 5, 6]
X = L * 4
Y = [L] * 4
X, Y
```

```python
L[1] = 666
X, Y
```

---

```python
L = [4, 5, 6]
Y = [list(L)] * 3
print(Y)
L[1] = 666
print(Y)
```

```python
Y[0][1] = 666
Y
```

```python
id(Y[0]), id(Y[1])
```

---

```python
L = [4, 5, 6]
Y = [list(L) for i in range(3)]
print(Y)
Y[0][1] = 666
print(Y)
```
