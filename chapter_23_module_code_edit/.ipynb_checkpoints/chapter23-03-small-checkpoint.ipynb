------ 重新启动 -----

```python
import sys
print(len(sys.modules), sys.modules.get('small', 666))
import small
print(len(sys.modules), sys.modules.get('small', 666))

print(small.x, small.y)
small.x = 2
small.y[0] = 42
print(small.x, small.y)
```

```python
from small import x, y

print(x, y)
```
