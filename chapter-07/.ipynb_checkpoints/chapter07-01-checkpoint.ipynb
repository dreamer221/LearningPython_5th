```python
import sys
import os

# Windows 系统
if sys.platform == 'win32':
    import winsound
    winsound.Beep(1000, 500)  # 频率1000Hz，持续500ms
    # 或播放系统提示音
#     winsound.MessageBeep()

# macOS 系统
elif sys.platform == 'darwin':
    os.system('afplay /System/Library/Sounds/Ping.aiff 2>/dev/null')

# Linux 系统
else:
    # 尝试使用 beep 命令（可能需要安装）
    os.system('echo -e "\a"')
    # 或使用系统声音
    os.system('paplay /usr/share/sounds/freedesktop/stereo/complete.oga 2>/dev/null')
```

```python
# \r \n区别和联系
# \r（Carriage Return，回车，ASCII 13）：把光标移到当前行的开头
# \n（Line Feed，换行，ASCII 10）：把光标移到下一行
```

```python
s = 'abcdefg\rABCD\nxyz'
s
```

```python
# 先打印"abcdefg"，回车回到行首，再打印"ABCD"覆盖了"abcd",再移到下一行打印
# 如果是写入文件,则不会覆盖,见下面的写入案例
print(s)
```
