## \__doc\__ 查看文档字符串

```python
def square(x):
    """
    fucntion document
    can we have your liver then?
    """
    return x ** 2
```

```python
# 函数的文档字符串
square.__doc__
```

```python
class Employee:
    """
    class documentation
    """
    pass
    
    def f1():
        """
        类的方法的document
        """
        pass
```

```python
# 类的文档字符串
Employee.__doc__
```

```python
# 类的方法的文档字符串
Employee.f1.__doc__
```

```python
import docstrings
```

```python
# 模块的文档字符串
docstrings.__doc__
```
