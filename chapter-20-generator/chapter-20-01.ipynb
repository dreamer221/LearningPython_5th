### 为什么要使用生成器函数

```python
def buildsquares(n):
    res = []
    for i in range(n):
        res.append(i ** 2)
    return res
```

```python
li = buildsquares(5)
for i in li:
    print(i, end=' : ')
```

```python
for x in [n ** 2 for n in range(5)]:
    print(x, end=' : ')
```

```python
for x in map(lambda n : n ** 2, range(5)):
    print(x, end=' : ')
```

## yield from

```python
def both(N):
    for i in range(N):
        yield i
    yield '--'
    for i in range(N):
        yield i ** 2
```

```python
b = both(5)
list(b)
```

```python
def both2(N):
    yield from range(N)
    yield '--'
    yield from (i ** 2 for i in range(N))
```

```python
b2 = both2(5)
list(b2)
```

#### 编写自己的map(func,..)

```python
def mymap1(func, *seqs):
    res = []
    for args in zip(*seqs):
        res.append(func(*args))
    return res
```

```python
mymap1(abs, [-2, -1, 0, 1, 2])
```

```python
mymap1(pow, [1, 2, 3], [2, 3, 4, 5])
```

```python
def mymap2(func, *seqs):
    return [func(*args) for args in zip(*seqs)]
```

```python
print(mymap2(abs, [-2, -1, 0, 1, 2]), 
      mymap2(pow, [1, 2, 3], [2, 3, 4, 5]) )
```

```python
def mymap3(func, *seqs):
    for args in zip(*seqs):
        yield func(*args)
```

```python
g3 = mymap3(abs, [-2, -1, 0, 1, 2])
next(g3), next(g3)
```

```python
def mymap4(func, *seqs):
    return (func(*args) for args in zip(*seqs))
```

```python
g4 = mymap3(abs, [-2, -1, 0, 1, 2])
next(g4)
```

```python
for i in g4:
    print(i)
```

###  请留意: 单遍迭代

```python
def myzip(*args):
    iters = map(iter, args)
    
    while iters:
        res = [next(i) for i in iters] 
        # iters元素都是生成器,但是他本身也是生成器, 
        # 上面迭代一次后, 虽然每个元素仍可以迭代,但是他自己已经是空了
        print(res, 'iters:', list(iters))
        yield tuple(res)
```

```python
list(myzip('abc', 'lmnop'))
```

```python
def myzip2(*args):
    iters = list(map(iter, args))
    while iters: # 也不对啊,这个永远是True,需要加异常判断
        try:
            res = [next(i) for i in iters] 
            print(res, iters)
            yield tuple(res)
        except StopIteration:
            break
```

```python
list(myzip2('abc', 'lmnop'))
```
