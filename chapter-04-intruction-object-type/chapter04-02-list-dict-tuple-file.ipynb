## How to Break the Flexibility of Code

```python
li = [1, 2, 3]
type(li), type(type(li))
```

```python
type(li) == type([])
```

```python
type(li) == list
```

```python
isinstance(li, list)
```

- 判断了类型，也就破坏了python的灵活性，所以并不建议使用这样的代码判断
- 不要限制在特性的类型上，要有多态的思想

## 用户定义的类

```python
class Worker:
    def __init__(self, name, pay):
        self.name = name
        self.pay = pay
        
    def lastName(self):
        return self.name.split()[-1]
    
    def giveRaise(self, percent):
        self.pay *= 1 + percent
        
w1 = Worker('Bob Smith', 50000)
w2 = Worker('Sue Jones', 60000)
```

```python
w1.lastName()
```

```python
w2.giveRaise(0.1)
w2.pay
```
